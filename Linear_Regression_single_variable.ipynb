This is a Liner Regression parctice notebook, which predicts the canada per capita income based on the year.
The model is trained on Linear Regression from scikit learn (sklearn) Library.

First import all the required libarry

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

Import the training dataset

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
df.rename(columns = {'per capita income (US$)' : 'perCapitaIncome'}, inplace = True)
df.head()
```

Plot the dataset in a scatter diagram

```python
%matplotlib inline
# plt.xlabel('Year')
# plt.ylabel('Per Capita Income (US$)')
plt.title('Scatter Plot Per Capita Income (US$) vs Year')
plt.scatter(df.year, df.perCapitaIncome, color = 'green')
plt.show()
```

```python
#create a Linearregression object
reg = linear_model.LinearRegression()
# Train the LinearRegression model bsaed on the labeled data
reg.fit(df[['year']], df.perCapitaIncome)
```

But the above process of training is NOT suitable for
everywhere. We have to break the dataset into ***x_train, x_test, y_train, y_test***.
***x_train and x_test*** is used to training the model.
***y_train and y_test*** is used to determine the accuracy of the model.
So here I divide the dataset into 2 part, one part will be independent variables and other part will be the target variable
Although only 1 independent variable is present here

```python
new_df = df.drop('perCapitaIncome', axis = 'columns')
new_df.head()
```

```python
perCapitaIncome = df.perCapitaIncome
perCapitaIncome.head()
```

```python
#Now train the model on the new spiltted data
model = linear_model.LinearRegression();
model.fit(new_df, perCapitaIncome)
```

```python
#Now predict the perCapitaIncome on the year 2025
model.predict([[2025]])
```
