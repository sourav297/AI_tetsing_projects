```python
# Select Individuals Columns
ind=homeless['individuals']
ind
```

```python
# Filter for rows where individuals is greater than 10000
ind_grt_10K=homeless[homeless['individuals']>10000]
ind_grt_10K
```

```python
# Create a Subset for rows in South Atlantic or Mid-Atlantic Regions By Categorical values
south_mid_alt=homeless[(homeless['region']=='South Atlantic') | (homeless['region']=='Mid-Atlantic')]
south_mid_alt
```

**Whatever We have Received May be Not Enough - So We Decided to Put a New Cols for betterment of our result - Sometimes this is called feature engineering**

```python
# Add total cols as sum of ind and family memebers
homeless['total']=homeless['individuals']+homeless['family_members']
```

```python
homeless.head()
```

```python
# Add p_ind cols as proportion of total that are ind
homeless['p_ind']=homeless['individuals']+homeless['total']
```

```python
homeless.head()
```

**Combo Attack**

**Which State has the highest number of homeless individuals per 10000 people in the state?**

```python
# Create In_per_10K cols as homeless ind per 10K state pop
homeless['ind_per_10K']=10000*homeless['individuals']/homeless['state_pop']
```

```python
homeless.head()
```

```python
# Subset rows for individuals ind_per_10K greater than 20
high_homeless=homeless[homeless['ind_per_10K']>20]
high_homeless
```

```python
# Line No 27 Can be Sorted too
high_homeless_sort=high_homeless.sort_values('ind_per_10K',ascending=True)
```
