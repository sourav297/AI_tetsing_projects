For multiple variable the Linear Regression model will be like
`y = m1*x1 + m2*x2 + m3*x3 + ... + mn*xn + c`
where m1, m2,..., mn are coefficients and x1, x2, ..., xn are input features

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
#resd the 'homeprices.csv' file
df = pd.read_csv('homeprices.csv')
df
```

There is a empty value in bedrooms column, So we have to first preprocess the data and for that I use here median to fill the empty or NaN field

```python
df.bedrooms.median()
```

```python
#Now fill this value to the empty fields in the dataset by fillna() method
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
#Split the dataset into input and output datas
new_df = df.drop('price', axis = 'columns')
new_df.head()
```

```python
price = df.price
price.head()
```

```python
#Now create a Linear Regression object and train the model
model = linear_model.LinearRegression()
#model.fit([['area', 'bedrooms', 'age']], df.price)  //This syntax can also be used
model.fit(new_df, price)
```

```python
#now print all the solpes or coefficients
model.coef_
```

```python
#now print the intercept
model.intercept_
```

```python
#now predict the price of a home with area = 3000, bedrooms = 3, age = 40
model.predict([[3000, 3, 40]])
```

```python
#validation of the prediction
# y = m1*x1 + m2*x2 + m3*x3 + c
y = 112.06244194*3000 + 23388.88007794*3 + (-3231.71790863*40) + 221323.00186540396
print(y)
```

```python
#now predict the price of a home with area = 2500, bedrooms = 4, age = 5
model.predict([[2500, 4, 5]])
```

```python
#now predict the price of a home with area = 4100, bedrooms = 6, age = 2
model.predict([[4100, 6, 2]])
```

Now build a Linear Regression model to predict the salary based on the 3 independent variable or input feature experience, test_score(out of 10) and interview_score(out of 10)...
First download the 'hiring.csv' file and then do rest of the work.

```python
dataset = pd.read_csv('hiring.csv')
dataset
```

Now preprocess the dataset

```python
#rename the 2nd and 3rd ans 4th column
dataset.rename(columns = {'test_score(out of 10)': 'test_score', 'interview_score(out of 10)' : 'interview_score', 'salary($)' : 'salary'}, inplace = True)
dataset
```

```python
#fill the empty cell in test_score column by the mean of that column
import math
test_mean = math.floor(dataset.test_score.mean())
print(test_mean)
dataset.test_score = dataset.test_score.fillna(test_mean)
dataset.experience = dataset.experience.fillna('zero')
dataset
```

```python
#convert the values in experience column to numeric value
!pip install word2number
from word2number import w2n
```

```python
#Now we can use the module word2number
dataset.experience = dataset.experience.astype(str).apply(w2n.word_to_num)
dataset
```

```python
#now split the dataset into input and output datas
#This is input part
new_dataset = dataset.drop('salary', axis = 'columns')
new_dataset
```

```python
#This is output part
salary = dataset.salary
salary
```

```python
#now create the object for Linear Regression and train the model
hiring_model = linear_model.LinearRegression()
hiring_model.fit(new_dataset, salary)
```

```python
#print the slopes or coefficients
hiring_model.coef_
```

```python
#print the intercept
hiring_model.intercept_
```

```python
#predict for the new input features
hiring_model.predict([[2, 9, 6]])
```
