```python
#read the csv file
df = pd.read_csv('titanic.csv')
df.head()
```

Here use only 'Survived', 'Pclass', 'Sex', 'Age', 'Fare' columns to build the model and 'Survived' is the target feature.

1.   List item
2.   List item



```python
df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis=1)
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
inputs.head()
```

```python
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
```

```python
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs.head()
```

```python
new_inputs = inputs.drop(['Sex'], axis='columns')
new_inputs.head()
```

```python
new_inputs.columns[new_inputs.isna().any()]
```

```python
new_inputs.Age = new_inputs.Age.fillna(new_inputs.Age.mean())
new_inputs.head()
```

```python
new_inputs.columns[new_inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_inputs,target,test_size=0.2, random_state=17)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#import the decision tree class and create a object of this class and then train the model
from sklearn import tree
my_model = tree.DecisionTreeClassifier()
```

```python
my_model.fit(X_train, y_train)
```

```python
my_model.score(X_test, y_test)
```

```python
my_model.predict([[1, 44.000000, 57.9792, 0]])  # 523 th row of X_train
```

```python
my_model.predict([[3, 47.000000, 14.5000, 0]])  # 132 th row of X_train
```





**Save Model to a file using python pickle**

```python
import pickle
```

```python
# save the model
with open ('model_pickle', 'wb') as file:
  pickle.dump(my_model, file)
```
